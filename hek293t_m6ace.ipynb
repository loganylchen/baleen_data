# Hek293t m6ace-seq ground truth 

## This file was from the author of the m6ace-seq

```python
raw_hek293t_df = pd.read_excel('hek293t_m6ace/m6ace_hek293t_fromAuthor.xlsx')
raw_hek293t_df['Chr']=raw_hek293t_df['Chr'].str.replace('chr','')
raw_hek293t_df['WT RML average'] = raw_hek293t_df.loc[:,['WT rep1 RML', 'WT rep2 RML','WT rep3 RML']].mean(axis=1)
raw_hek293t_df['Mettl3-KO RML average'] = raw_hek293t_df.loc[:,['Mettl3-KO rep1 RML', 'Mettl3-KO rep2 RML','Mettl3-KO rep3 RML']].mean(axis=1)
raw_hek293t_df['WT-KO'] = raw_hek293t_df['WT RML average']/raw_hek293t_df['Mettl3-KO RML average']
raw_hek293t_df['mettl3-m6a'] = 0
raw_hek293t_df.loc[raw_hek293t_df['WT-KO'] >=4,'mettl3-m6a'] = 1
raw_hek293t_df.head()
```

## This is from the m6ace-seq paper supplementary data 4

```python
m6aceseq_hek293t_df = pd.read_excel('hek293t_m6ace/m6ace_hek293t_fromPaper.xlsx')
m6aceseq_hek293t_df['Chr']=m6aceseq_hek293t_df['Chr'].str.replace('chr','')
m6aceseq_hek293t_df['WT RML average'] = m6aceseq_hek293t_df.loc[:,['WT rep1 RML', 'WT rep2 RML','WT rep3 RML']].mean(axis=1)
m6aceseq_hek293t_df['Mettl3-KO RML average'] = m6aceseq_hek293t_df.loc[:,['Mettl3-KO rep1 RML', 'Mettl3-KO rep2 RML','Mettl3-KO rep3 RML']].mean(axis=1)
m6aceseq_hek293t_df['WT-KO'] = m6aceseq_hek293t_df['WT RML average']/m6aceseq_hek293t_df['Mettl3-KO RML average']
m6aceseq_hek293t_df['mettl3-m6a'] = 0
m6aceseq_hek293t_df.loc[m6aceseq_hek293t_df['WT-KO'] >=4,'mettl3-m6a'] = 1
m6aceseq_hek293t_df.head()
```

```python
print(xpore_hek293t_df.shape)
print(m6aceseq_hek293t_df.shape)
print(m6aceseq_hek293t_df.merge(xpore_hek293t_df,on=['Chr','Start','End','Strand'],how='outer').shape)

# So the supplementary file was the same one as the xpore one
```
